### Run Models With Train/Test Split and RMSE

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
# Define a function that accepts a list of features and returns testing RMSE.
def train_test_rmse(df, feature_cols):
    X = df[feature_cols]
    y = df.Absent_time
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123) #Use default split, set a ramdon_state value for reproduction
    
    linreg = LinearRegression()
    linreg.fit(X_train, y_train)
    
    y_pred = linreg.predict(X_test)
    return np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```

## Decision Tree

#### Model 1： Only includes all the dummy values for column Reason

```python
#Define my features
feature_cols = ['Reason_1', 'Reason_2',
       'Reason_4', 'Reason_5', 'Reason_6', 'Reason_7', 'Reason_8', 'Reason_9',
       'Reason_10', 'Reason_11', 'Reason_12', 'Reason_13', 'Reason_14',
       'Reason_15', 'Reason_16', 'Reason_17', 'Reason_18', 'Reason_19',
       'Reason_21', 'Reason_22', 'Reason_23', 'Reason_24', 'Reason_25',
       'Reason_26', 'Reason_27', 'Reason_28']
# Define X and y.
X = reason_dummies[feature_cols]
y = reason_dummies.Absent_time
```

```python
from sklearn.tree import DecisionTreeRegressor

# Grow each tree deep.
treereg = DecisionTreeRegressor(max_depth=None, random_state=123)

#use cross-validation to find optimal depth of a decision tree
# List of values to try for max_depth:
max_depth_range = list(range(1, 30))

# List to store the average RMSE for each value of max_depth:
RMSE_scores = []

# Use 10-fold cross-validation with each value of max_depth.
from sklearn.model_selection import cross_val_score
for depth in max_depth_range:
    treereg = DecisionTreeRegressor(max_depth=depth, random_state=123)
    MSE_scores = cross_val_score(treereg, X, y, cv=10, scoring='neg_mean_squared_error')
    RMSE_scores.append(np.mean(np.sqrt(-MSE_scores)))
```

```python
print(RMSE_scores) #print out RMSE_scores
```

### When tree depth is 26, we get the lowest RMSE score 12.643718. 

```python
# Looking for optimum max depth of the tree - when depth is 26, RMSE_scores has the lowest number 
# Plot max_depth (x-axis) versus RMSE (y-axis).
max_depth_range = list(range(1, 30)) 
plt.plot(max_depth_range, RMSE_scores);
plt.xlabel('max_depth');
plt.ylabel('RMSE (lower is better)');
```

```python
treereg = DecisionTreeRegressor(max_depth=26, random_state=123)
treereg.fit(X, y)
```

```python
# Compute feature importances in decision tree. We see even the feature with highest importance rate is still very low. 
pd.DataFrame({'feature':feature_cols, 'importance':treereg.feature_importances_}).sort_values(by='importance')
```

```python
#Get the RMSE score by using Train/Test split. The RMSE scoree is still bigger than Null RMSE
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123) #Use default split, set a ramdon_state value for reproduction
    
treereg = DecisionTreeRegressor(max_depth=26, random_state=123)
treereg.fit(X_train, y_train)
    
y_pred = treereg.predict(X_test)
RMSE_train_test = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```

```python
print(RMSE_train_test)
```

#### Reducing X to its Most Important Features

```python
X.shape # Intotal, X has 36 columns (features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 123)
# Fit the model on only the train data
rfreg = RandomForestRegressor(n_estimators=150, max_features=5, oob_score=True, random_state=123)
rfreg.fit(X_train, y_train)
```

```python
from sklearn.feature_selection import SelectFromModel

print(SelectFromModel(rfreg, threshold='mean', prefit=True).transform(X_train).shape)
print(SelectFromModel(rfreg, threshold='median', prefit=True).transform(X_train).shape)
```

```python
# Create a new feature matrix that only includes important features.

X_important =  SelectFromModel(rfreg, threshold='mean', prefit=True).transform(X_test)
```

```python
# Check the RMSE for a random forest that only includes important features.
rfreg = RandomForestRegressor(n_estimators=150, max_features=3, random_state=123)

scores = cross_val_score(rfreg, X_important, y_test, cv=10, scoring='neg_mean_squared_error')
np.mean(np.sqrt(-scores))
```

#### We are getting a better result !!!

#### Let's try to optimize the model even better by tuning n_estimators

```python
# List of values to try for n_estimators:
estimator_range = list(range(10, 300, 10))

# List to store the average RMSE for each value of n_estimators:
RMSE_scores = []

feature_cols_opt = ['Reason_23', 'Son', 'Body_mass', 'Reason_12', 'Reason_11', 'Service_time', 'Reason_13', 'Trans_expense', 'Reason_19', 'Reason_9', 'Distance', 'Age', 'Work_load']
X_opt = reason_dummies[feature_cols_opt]
# Use five-fold cross-validation with each value of n_estimators (Warning: Slow!).
for estimator in estimator_range:
    rfreg = RandomForestRegressor(n_estimators=estimator, random_state=123)
    MSE_scores = cross_val_score(rfreg, X_opt, y, cv=5, scoring='neg_mean_squared_error')
    RMSE_scores.append(np.mean(np.sqrt(-MSE_scores)))
print(RMSE_scores)
```

```python
plt.plot(estimator_range, RMSE_scores);

plt.xlabel('n_estimators');
plt.ylabel('RMSE (lower is better)');
```

```python
sorted(zip(RMSE_scores, estimator_range))[0] # Let's choose 300 to be the n_estimators 
```

#### Tuning max_features

```python
# Altered n_estimators based on optimal finding above
# Now let's look for the optimum number of maximum features at each split

# List of values to try for max_features:
feature_range = list(range(1, len(feature_cols_opt)+1))

# List to store the average RMSE for each value of max_features:
RMSE_scores = []

# Used 5-fold cross-validation with each value of max_features.
for feature in feature_range:
    rfreg = RandomForestRegressor(n_estimators=300, max_features=feature, random_state=123)
    MSE_scores = cross_val_score(rfreg, X_opt, y, cv=5, scoring='neg_mean_squared_error')
    RMSE_scores.append(np.mean(np.sqrt(-MSE_scores)))
print(RMSE_scores)
```

```python
# Plot max_features (x-axis) versus RMSE (y-axis).

plt.plot(feature_range, RMSE_scores);

plt.xlabel('max_features');
plt.ylabel('RMSE (lower is better)');
```

```python
sorted(zip(RMSE_scores, feature_range))[0]
```

#### We can tell when max_features equals 6, the RMSE has the smallest number

### Let's re-train out data by using the optimized n_estimators and max_features values.

```python
from sklearn.model_selection import train_test_split

#restating X for clarity and designating y
X_opt = reason_dummies[feature_cols_opt]
y = reason_dummies.Absent_time

X_train, X_test, y_train, y_test = train_test_split(X_opt,y, random_state = 123)
rfreg = RandomForestRegressor(n_estimators=300, max_features=5, random_state=123)
rfreg.fit(X_train, y_train)
```

```python
#Check the RMSE for the optimized RF model on the new test data

scores = cross_val_score(rfreg, X_test, y_test, cv=10, scoring='neg_mean_squared_error')
np.mean(np.sqrt(-scores))
```
