### Dummify the column Reason

```python
reason_dummies = pd.get_dummies(absent_new.Reason, prefix='Reason')
```

```python
reason_dummies.sample(n=5, random_state=1) # Print five random rows.
```

```python
reason_dummies.drop(reason_dummies.columns[2], axis=1, inplace=True) #Drop the third row (because from the box plot we drew earlier, the third row has simple value, I think it serves a good baseline)
```

```python
reason_dummies.sample(n=5, random_state=1) #Print five random rows again (the previous first row is dropped)
```

#### Combine two DataFrame into one

```python
# Concatenate the original DataFrame and the dummy DataFrame (axis=0 means rows, axis=1 means columns).
reason_dummies = pd.concat([absent_new, reason_dummies], axis=1)

# Print five random rows.
reason_dummies.sample(n=5, random_state=1)
```

### Run Linear Regreassion with the new DataFrame with dummy variables included

```python
# Include all the dummy variables for column Reason in the model.
feature_cols = ['Reason_1', 'Reason_2',
       'Reason_4', 'Reason_5', 'Reason_6', 'Reason_7', 'Reason_8', 'Reason_9',
       'Reason_10', 'Reason_11', 'Reason_12', 'Reason_13', 'Reason_14',
       'Reason_15', 'Reason_16', 'Reason_17', 'Reason_18', 'Reason_19',
       'Reason_21', 'Reason_22', 'Reason_23', 'Reason_24', 'Reason_25',
       'Reason_26', 'Reason_27', 'Reason_28']
X = reason_dummies[feature_cols]
y = reason_dummies.Absent_time

# Import, instantiate, fit.
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X, y)

list(zip(feature_cols, linreg.coef_))
```

### Run Models With Train/Test Split and RMSE

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
# Define a function that accepts a list of features and returns testing RMSE.
def train_test_rmse(df, feature_cols):
    X = df[feature_cols]
    y = df.Absent_time
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123) #Use default split, set a ramdon_state value for reproduction
    
    linreg = LinearRegression()
    linreg.fit(X_train, y_train)
    
    y_pred = linreg.predict(X_test)
    return np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```

#### Model 1： Only includes all the dummy values for column Reason

```python
print(train_test_rmse(reason_dummies, ['Reason_1', 'Reason_2',
       'Reason_4', 'Reason_5', 'Reason_6', 'Reason_7', 'Reason_8', 'Reason_9',
       'Reason_10', 'Reason_11', 'Reason_12', 'Reason_13', 'Reason_14',
       'Reason_15', 'Reason_16', 'Reason_17', 'Reason_18', 'Reason_19',
       'Reason_21', 'Reason_22', 'Reason_23', 'Reason_24', 'Reason_25',
       'Reason_26', 'Reason_27', 'Reason_28']))
```

#### Model 2: Only includes the three features with the lagest corr with Absent time.

```python
print(train_test_rmse(reason_dummies, ['Son', 'Distance', 'Age']))
```

#### Model 3: Only includes the feature with the lagest corr with Absent time.

```python
print(train_test_rmse(reason_dummies, ['Son']))
```

#### Model 4: Includes all the dummy values fro column Reason and also other three features with the lagest corr with Absent time.

#### Comparing Testing RMSE With Null RMSE

```python
# Split X and y into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123)

# Create a NumPy array with the same shape as y_test.
y_null = np.zeros_like(y_test, dtype=float)

# Fill the array with the mean value of y_test.
y_null.fill(y_test.mean())
y_null
```

```python
# Compute null RMSE.
np.sqrt(metrics.mean_squared_error(y_test, y_null))
```

### Observation:
Even the best model from our 4 models has a higher RMSE than the Null RMSE. The linear regression is definietly not for prediciting Absent time based on the features we have in our dataset.
